```python
import pandas as pd

movies = {
    "Movie": [
        "Avengers",
        "Iron Man",
        "Thor",
        "Batman",
        "Superman"
    ],
    "Genre": [
        "Action Superhero",
        "Action Superhero",
        "Action Fantasy",
        "Action Dark",
        "Action Hero"
    ]
}

df = pd.DataFrame(movies)

df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer()

matrix = cv.fit_transform(df["Genre"])

similarity = cosine_similarity(matrix)

print(similarity)
```

```python
def recommend(movie):
    movie_index = df[df["Movie"] == movie].index[0]

    distances = similarity[movie_index]

    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:4]

    print("Recommended Movies for", movie, ":")

    for i in movie_list:
        print(df.iloc[i[0]].Movie)
```

```python
recommend("Avengers")
```
